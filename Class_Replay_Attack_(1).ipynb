```python
# STEP 1: Unzip dataset
import zipfile, os, shutil, random
from pathlib import Path

zip_path = "/content/2_Class_Relay_Attack.zip"
extract_path = "/content"
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

original_dataset_dir = "/content/2_Class_Relay_Attack/TRAIN"
base_dir = "/content/split_data"
splits = ['train', 'val', 'test']
classes = ['GN', 'IM']

# STEP 2: Clean previous split directory if exists
if os.path.exists(base_dir):
    shutil.rmtree(base_dir)

# STEP 3: Create folder structure
for split in splits:
    for cls in classes:
        os.makedirs(os.path.join(base_dir, split, cls), exist_ok=True)

# STEP 4: Split data from each class folder (70/15/15)
def split_and_copy(class_name):
    src = os.path.join(original_dataset_dir, class_name)
    all_images = [f for f in os.listdir(src) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
    random.shuffle(all_images)

    total = len(all_images)
    n_train = int(0.7 * total)
    n_val = int(0.15 * total)
    n_test = total - n_train - n_val

    splits_map = {
        'train': all_images[:n_train],
        'val': all_images[n_train:n_train + n_val],
        'test': all_images[n_train + n_val:]
    }

    for split, files in splits_map.items():
        for fname in files:
            shutil.copy(os.path.join(src, fname), os.path.join(base_dir, split, class_name))

for cls in classes:
    split_and_copy(cls)

print("✅ Dataset split: 70% train, 15% val, 15% test for each class.")

# STEP 5: Load dataset
from tensorflow.keras.preprocessing.image import ImageDataGenerator

img_height, img_width = 128, 128
batch_size = 16

datagen = ImageDataGenerator(rescale=1.0 / 255)

train_gen = datagen.flow_from_directory(
    os.path.join(base_dir, 'train'),
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary'
)

val_gen = datagen.flow_from_directory(
    os.path.join(base_dir, 'val'),
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary'
)

test_gen = datagen.flow_from_directory(
    os.path.join(base_dir, 'test'),
    target_size=(img_height, img_width),
    batch_size=1,
    class_mode='binary',
    shuffle=False
)

# STEP 6: Build CNN model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(img_height, img_width, 3)),
    MaxPooling2D(2, 2),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Conv2D(256, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Conv2D(512, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Flatten(),
    Dropout(0.5),
    Dense(256, activation='relu'),
    Dense(1, activation='sigmoid')  # Binary classification
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

# STEP 7: Train model
history = model.fit(
    train_gen,
    epochs=30,
    validation_data=val_gen
)

# STEP 8: Plot Accuracy & Loss
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.title("Model Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title("Model Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()

plt.tight_layout()
plt.show()

# STEP 9: Evaluation - Confusion Matrix & Classification Report
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay

y_true = test_gen.classes
y_probs = model.predict(test_gen)
y_pred = (y_probs > 0.5).astype(int).reshape(-1)

# Print classification report
target_names = list(test_gen.class_indices.keys())
print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=target_names))

# STEP 10: Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=target_names)
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.show()

# STEP 11: Evaluate on test set
test_loss, test_accuracy = model.evaluate(test_gen)
print(f"\n✅ Test Accuracy: {test_accuracy * 100:.2f}%")

# STEP 12: Save model
model.save("/content/face_antispoofing_model.h5")
print("✅ Model saved at /content/face_antispoofing_model.h5")

```

```python
import numpy as np
import matplotlib.pyplot as plt
import lime
from lime import lime_image
from skimage.segmentation import mark_boundaries
from tensorflow.keras.preprocessing import image
import tensorflow as tf
import os

def analyze_activation(mask):
    mask = (mask - np.min(mask)) / (np.max(mask) - np.min(mask))
    suspicious_regions = mask > 0.5
    total_pixels = mask.size
    suspicious_pixels = np.sum(suspicious_regions)
    suspicious_percent = (suspicious_pixels / total_pixels) * 100

    h, w = mask.shape[:2]
    border_thickness_h = int(h * 0.15)
    border_thickness_w = int(w * 0.15)

    border_mask = np.zeros_like(mask, dtype=bool)
    border_mask[:border_thickness_h, :] = True
    border_mask[-border_thickness_h:, :] = True
    border_mask[:, :border_thickness_w] = True
    border_mask[:, -border_thickness_w:] = True

    suspicious_border = np.sum(np.logical_and(suspicious_regions, border_mask))
    suspicious_center = np.sum(np.logical_and(suspicious_regions, ~border_mask))

    border_percent = (suspicious_border / suspicious_pixels) * 100 if suspicious_pixels > 0 else 0
    center_percent = (suspicious_center / suspicious_pixels) * 100 if suspicious_pixels > 0 else 0

    return suspicious_percent, border_percent, center_percent

def generate_dynamic_explanation(predicted_class, suspicious_percent, border_percent, center_percent):
    if predicted_class == 0:  # Genuine
        if suspicious_percent < 10:
            reason = ("Minimal suspicious activation detected. "
                      "Facial patterns appear consistent with genuine behavior.")
        elif suspicious_percent < 25:
            reason = ("Some minor suspicious regions detected, "
                      "but overall facial texture and structure support real face classification.")
        else:
            reason = ("Moderate suspicious activation noted, especially at edges. "
                      "However, central facial regions remain consistent with genuine patterns.")
    else:  # Imposter
        if border_percent > 50 and suspicious_percent > 20:
            reason = (" Strong suspicious activation around image borders detected, "
                      "indicating possible spoofing artifacts like printed edges or screen reflections.")
        elif center_percent > 50:
            reason = (" Suspicious activations concentrated on facial areas (eyes, nose, mouth), "
                      "suggesting tampering or replay attacks.")
        else:
            reason = (" Distributed suspicious activations detected across the face and edges, "
                      "raising concerns about image authenticity.")
    return reason

def predict_and_explain(img_path, model, save_output=False, output_dir="/content/output"):
    img = image.load_img(img_path, target_size=(128, 128))
    img_array = image.img_to_array(img) / 255.0
    img_batch = np.expand_dims(img_array, axis=0)

    prediction = model.predict(img_batch)[0][0]
    predicted_class = 0 if prediction < 0.5 else 1
    label = "Genuine" if predicted_class == 0 else "Imposter"

    # Explain with LIME
    explainer = lime_image.LimeImageExplainer()
    explanation = explainer.explain_instance(
        img_array.astype('double'),
        model.predict,
        top_labels=1,
        hide_color=0,
        num_samples=1000
    )

    temp, mask = explanation.get_image_and_mask(
        explanation.top_labels[0],
        positive_only=False,
        num_features=5,
        hide_rest=False
    )

    suspicious_percent, border_percent, center_percent = analyze_activation(mask)
    reason = generate_dynamic_explanation(predicted_class, suspicious_percent, border_percent, center_percent)

    # Color Coding
    from skimage.color import label2rgb
    color = (0, 1, 0) if predicted_class == 0 else (1, 0, 0)  # Green for genuine, Red for imposter
    temp_colored = label2rgb(mask, temp, colors=[color], alpha=0.6, bg_label=0)

    # --- Plot ---
    fig, ax = plt.subplots(1, 2, figsize=(14, 5))

    # Set Explanation Text Color
    text_color = 'green' if predicted_class == 0 else 'red'
    # Original Image
    ax[0].imshow(img_array)
    ax[0].set_title(f"Original Image\nPrediction: {label} ({prediction:.4f})", fontsize=14, color=text_color, fontweight='bold', loc='center')
    ax[0].axis('off')

    # Explanation Image + Text
    ax[1].imshow(temp_colored)
    ax[1].axis('off')


    ax[1].set_title(reason, fontsize=14, color=text_color, loc='left', fontweight='bold',wrap = "True")

    plt.tight_layout()

    # Save output
    if save_output:
        os.makedirs(output_dir, exist_ok=True)
        filename = os.path.basename(img_path).split('.')[0]
        plt.savefig(os.path.join(output_dir, f"{filename}_{label}.png"))
        print(f"Saved to {os.path.join(output_dir, f'{filename}_{label}.png')}")

    plt.show()
```
